Final Merge with Composer

```python
import pandas as pd

# Load the datasets
enriched_data = pd.read_excel("enriched_data_with_composer.xlsx")
movies_ratings_enriched = pd.read_csv("movies_ratings_enriched.csv")

# Rename "Movie Title" column to "Title"
enriched_data.rename(columns={'Movie Title': 'Title'}, inplace=True)

# Merge datasets using left join on "Title"
merged_data = pd.merge(movies_ratings_enriched, enriched_data, on='Title', how='left')

# Save the merged data as "final.csv"
merged_data.to_csv("final.csv", index=False)
```
